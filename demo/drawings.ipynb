```python
from numpy import array,sin,cos,pi
%matplotlib agg
from matplotlib.pyplot import close,subplots
from matplotlib.patches import FancyArrowPatch
```

## pendulum.png

```python
bbox = array(((-.25,1.4),(-2.5,.4)))
l = 2.5; θ = pi/6

bbox_r = bbox[:,1]-bbox[:,0]; bbox_r = bbox_r[0]/bbox_r[1]
x,y = l*sin(θ),-l*cos(θ)

close('all')
fig,ax = subplots(
  subplot_kw=dict(aspect='equal',xlim=bbox[0],ylim=bbox[1],xticks=(),yticks=(),frame_on=False),
  figsize=(3*bbox_r,3),tight_layout=dict(pad=0),facecolor='w')
ax.axhline(0,c='k',ls='--',lw=1)
ax.axvline(0,c='k',ls='--',lw=1)
ax.add_artist(FancyArrowPatch((.25,0),(0,.25),ls='--',arrowstyle='->',mutation_scale=10,connectionstyle='arc3,rad=.39'))
ax.arrow(0,0,0,-.5,fc='g',ec='none',width=.04)

ax.scatter((0,x),(0,y),s=(30,300),c=('k','r'),zorder=10)
ax.plot((0,x),(0,y),lw=2,c='k')
ax.add_artist(FancyArrowPatch((0,-1),(sin(θ),-cos(θ)),arrowstyle='->',mutation_scale=10,connectionstyle='arc3,rad=.25'))
ax.annotate(r'$\theta$',(sin(θ/2),-cos(θ/2)),xytext=(-3,-13),textcoords='offset points')
ax.annotate(r'$m$',(x,y),xytext=(5,-15),textcoords='offset points')
ax.annotate(r'$g$',(0,-.5),xytext=(-10,10),textcoords='offset points')
fig.savefig('pendulum.png')
fig
```

## pendulum-sliding.png

```python
bbox = array(((-.25,2.),(-2.5,.4)))
l = 2.5; θ = pi/6; xc = .5

bbox_r = bbox[:,1]-bbox[:,0]; bbox_r = bbox_r[0]/bbox_r[1]
x,y = xc+l*sin(θ),-l*cos(θ)

close('all')
fig,ax = subplots(
  subplot_kw=dict(aspect='equal',xlim=bbox[0],ylim=bbox[1],xticks=(),yticks=(),frame_on=False),
  figsize=(3*bbox_r,3),tight_layout=dict(pad=0),facecolor='w')
ax.axhline(0,c='k',ls='--',lw=1)
ax.axvline(0,c='k',ls='--',lw=1)
ax.add_artist(FancyArrowPatch((.25,0),(0,.25),ls='--',arrowstyle='->',mutation_scale=10,connectionstyle='arc3,rad=.39'))
ax.arrow(0,0,0,-.5,fc='g',ec='none',width=.04)
ax.arrow(1.2,0,.5,0,fc='g',ec='none',width=.04)
ax.plot((xc,xc),(.25,-1.2),c='k',ls='--',lw=1)

ax.scatter((xc,x),(0,y),s=(30,300),c=('k','r'),zorder=10)
ax.plot((xc,x),(0,y),lw=2,c='k')
ax.add_artist(FancyArrowPatch((xc,-1),(xc+sin(θ),-cos(θ)),arrowstyle='->',mutation_scale=10,connectionstyle='arc3,rad=.25'))
ax.annotate(r'$\theta$',(xc+sin(θ/2),-cos(θ/2)),xytext=(-3,-13),textcoords='offset points')
ax.annotate(r'$m$',(x,y),xytext=(5,-15),textcoords='offset points')
ax.annotate(r'$m_o$',(xc,0),xytext=(-15,5),textcoords='offset points')
ax.annotate(r'$g$',(0,-.5),xytext=(-10,10),textcoords='offset points')
ax.annotate(r'$\zeta(t)$',(1.5,0),xytext=(0,10),textcoords='offset points')
fig.savefig('pendulum-sliding.png')
fig
```

## pendulum-double.png

```python
bbox = array(((-.25,2),(-3,.4)))
l1 = 2.5; l2=.7; θ1 = pi/6; θ2 = -2*pi/3

bbox_r = bbox[:,1]-bbox[:,0]; bbox_r = bbox_r[0]/bbox_r[1]
x1,y1,x2,y2 = l1*sin(θ1),-l1*cos(θ1),l1*sin(θ1)+l2*sin(θ2),-l1*cos(θ1)-l2*cos(θ2)

close('all')
fig,ax = subplots(
  subplot_kw=dict(aspect='equal',xlim=bbox[0],ylim=bbox[1],xticks=(),yticks=(),frame_on=False),
  figsize=(3*bbox_r,3),tight_layout=dict(pad=0))
ax.axhline(0,c='k',ls='--',lw=1)
ax.axvline(0,c='k',ls='--',lw=1)
ax.add_artist(FancyArrowPatch((.25,0),(0,.25),ls='--',arrowstyle='->',mutation_scale=10,connectionstyle='arc3,rad=.39'))
ax.arrow(0,0,0,-.5,fc='g',ec='none',width=.04)
ax.plot((x1,x1),(y1-.7,y1+.3),c='k',ls='--',lw=1)

ax.scatter((0,x1,x2),(0,y1,y2),s=(30,300,80),c=('k','r','b'),zorder=10)
ax.plot((0,x1,x2),(0,y1,y2),lw=2,c='k')
ax.add_artist(FancyArrowPatch((0,-1),(sin(θ1),-cos(θ1)),arrowstyle='->',mutation_scale=10,connectionstyle='arc3,rad=.25'))
ax.add_artist(FancyArrowPatch((x1,y1-.4),(x1+.4*sin(θ2),y1-.4*cos(θ2)),arrowstyle='->',mutation_scale=10,connectionstyle='arc3,rad=-.4'))
ax.annotate(r'$\theta_1$',(sin(θ1/2),-cos(θ1/2)),xytext=(-3,-13),textcoords='offset points')
ax.annotate(r'$\theta_2$',(x1+.4*sin(θ2/2),y1-.4*cos(θ2/2)),xytext=(-6,-6),textcoords='offset points')
ax.annotate(r'$m_1$',(x1,y1),xytext=(5,-15),textcoords='offset points')
ax.annotate(r'$m_2$',(x2,y2),xytext=(-17,7),textcoords='offset points')
ax.annotate(r'$g$',(0,-.5),xytext=(-10,10),textcoords='offset points')
fig.savefig('pendulum-double.png')
fig
```

## G2body.png

```python
bbox = array(((-1,3.),(-.5,2)))
θ = pi/6; r = 3

bbox_r = bbox[:,1]-bbox[:,0]; bbox_r = bbox_r[0]/bbox_r[1]
x1,y1 = r*cos(θ),r*sin(θ)
x2,y2 = -.2*x1,-.2*y1

close('all')
fig,ax = subplots(
  subplot_kw=dict(aspect='equal',xlim=bbox[0],ylim=bbox[1],xticks=(),yticks=(),frame_on=False),
  figsize=(3*bbox_r,3),tight_layout=dict(pad=0))
ax.axhline(0,c='k',ls='--',lw=1)
ax.axvline(0,c='k',ls='--',lw=1)
ax.add_artist(FancyArrowPatch((.25,0),(0,.25),ls='--',arrowstyle='->',mutation_scale=10,connectionstyle='arc3,rad=.39'))

ax.scatter((x1,x2),(y1,y2),c=('r','b'),s=(50,300))
ax.plot((x1,x2),(y1,y2),ls=':',c='k')
ax.add_artist(FancyArrowPatch((1,0),(cos(θ),sin(θ)),arrowstyle='->',mutation_scale=10,connectionstyle='arc3,rad=.25'))
ax.arrow(x1,y1,-.25*x1,-.25*y1,fc='g',ec='none',width=.04)
ax.annotate(r'$\theta$',(cos(θ/2),sin(θ/2)),xytext=(3,0),textcoords='offset points')
ax.annotate(r'$m_1$',(x1,y1),xytext=(5,0),textcoords='offset points')
ax.annotate(r'$m_2$',(x2,y2),xytext=(-25,-10),textcoords='offset points')
ax.annotate(r'$F$',(.75*x1,.75*y1),xytext=(-5,5),textcoords='offset points')
fig.savefig('G2body.png')
fig
```
