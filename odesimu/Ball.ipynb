# Acceleration controller

Here, the system is a point body (ball) of mass $1$ controlled by a force which may depend on the state of the body (control loop).

The equations of the dynamics are given by

\begin{eqnarray*}
\left(\begin{array}{l}\ddot{x}\\ \ddot{y}\end{array}\right) & = & \gamma(t)
\end{eqnarray*}

$\gamma$ is assumed piecewise constant, with changing points at regular intervals (the control rate). At each changing points, the value of $\gamma$ is acquired from a controller, to which the current state of the system ($x,y,\dot{x},\dot{y}$) is communicated. The controller can thus keep a history of its observations, and base its decisions on it.

## Ball class

```python
#----------------------------------------------------------------------------------------------------
class Ball (System):
#----------------------------------------------------------------------------------------------------

  #integrator = dict(name='dopri5')
  # don't use the default intergrator lsoda, as it uses forward looking (not consistent with control)
  integrator = dict(name='dop853')

  def __init__(self,control):
    """
:param control: (lateral) acceleration, in m/s^2
    """
    self.control = control
    def main(t,state):
      x,y,dx,dy = state
      ddx,ddy = control(t)
      return array((dx,dy,ddx,ddy))
    self.main = main
    def fordisplay(state):
      x,y,dx,dy = state
      live = x,y
      return live,live
    self.fordisplay = fordisplay
    
  def runstep(self,ctrlrate=None,ini=None,**ka):
    self.control.reset(ini)
    t0 = T = 1/ctrlrate
    for t,y in super(Ball,self).runstep(ini=ini,**ka):
      if t>t0: self.control.update(t,y); t0 += T
      yield t,y

  def display(self,ax,bounds=None,**ka):
    ax.scatter((0,),(0,),c='k',marker='o')
    diag_s = ax.scatter((),(),c='b',marker='o')
    tail_l, = ax.plot((),(),'y')
    ax.set_title(r'Controlled quantity: acceleration (force)')
    ax.set_xlim(bounds[0][0],bounds[1][0])
    ax.set_ylim(bounds[0][1],bounds[1][1])
    def disp(t,live,tail):
      diag_s.set_offsets((live,))
      tail_l.set_data(tail[:,0],tail[:,1])
    return super(Ball,self).display(ax,disp,**ka)

  @staticmethod
  def makestate(x=0.,y=0.,dx=0.,dy=0.): return array((x,y,dx,dy))
```

Launcher
========

System parameters:

* `ini/x,y,dx,dy`: initial position and speed of the ball (in m and m.sec$^{-1}$)

* `ctrlrate`: control rate (in sec$^{-1}$)

* `srate`: simulation rate (in frames.sec$^{-1}$)

* `taild`: duration of shadow, ie. trace of previous states (in sec)

Controller parameters:

* `N`: size of the cache

* `target`: observed version of the target position function

* `obsf`: observation function mapping a state into an observed value

PID controller parameters:

* `gP,gI,gD`: Proportional, Integration, Derivative gains

```python
logger.setLevel(logging.WARNING)

period = 15
def target_still(t,spot=array((7.,5.))): return spot
def target_circle(t,R=4.,omega=2*pi/period):
  p = omega*t
  return R*hstack((sin(p),cos(p)))
def target_cycloid(t,R=4.,r=2.,l=-5.,omega=2*pi/period):
  p = omega*t
  return hstack((
    R*sin(p)+r*sin(l*p),
    R*cos(p)+r*cos(l*p),
    ))

#target = target_still
#target = buffered(T=period,N=int(period*100))(target_circle)
target = buffered(T=period,N=int(period*100))(partial(target_cycloid,l=5))

control = PIDController(
  N=2,
  gP=3.,
  #gP=.5,
  #gI=1.,
  gD=5.,
  #gD=7.,
  target=blurred(0.,shape=(2,))(target),
  observe=(lambda state=zeros((4,)):state[:2]),
  )
syst = Ball(control)
syst.launch(
  ini=syst.makestate(
    x=0.,
    y=0.,
    dx=10.,
    dy=0.,
    ),
  ctrlrate=5.,
  bounds=((-8.,-8.),(8.,8.)),
  hooks=(partial(point_display,f=target,marker='o',s=40),partial(controlLogger,logger=logger)),
  maxtime=infty,
  srate=25.,
  taild=period+5.,
  animate=dict(
    repeat=False,
    ),
  )
```
